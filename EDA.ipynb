# Global Capacity
Lets take a look at what the global capacity is for renewable energy, its growth, and how much of it is being used. First, lets just figure out what the capacity is for each country and how it has grown over time.

```python
# Creating a copy of the df that has the information that is needed
gc = df.drop(['ObjectId', 'ISO2', 'ISO3', 'Energy_Type', 'Source', 'CTS_Name', 'CTS_Code', 'CTS_Full_Descriptor', 'F2000', 'F2001',
             'F2002','F2003','F2004','F2005','F2006','F2007','F2008','F2009','F2010','F2011','F2012','F2013','F2014','F2015',
             'F2016','F2017','F2018','F2019','F2020','F2021','F2023',], axis=1)
gc.info()
```

```python
# replacing all the null values with 0
gc['F2022'].fillna(0)
gc.info()
```

```python
# I am going to need to iterate through a list of all the countries, sum there renewable production for 2022 (which appears to have more complete numbers than 2023).
country_list = gc['Country'].unique()
countru_list = country_list.tolist()
country_list
```

```python
print(len(country_list))
```

```python
# And a list of all the renewable column information
renewable_list = gc['Technology'].unique()
renewable_list = renewable_list.tolist()
renewable_list.remove('Fossil fuels')
renewable_list
```

```python
gc.head(10)
```

```python
gc = gc[gc['Indicator'] != 'Electricity Installed Capacity']
gc.tail(40)
```

```python

# List of renewable energy technologies
renewable_technologies = ['Bioenergy', 'Hydropower (excl. Pumped Storage)', 'Solar energy', 'Wind energy']

# 1. Filter the DataFrame to only include renewable energy technologies
df_renewables = gc[gc['Technology'].isin(renewable_technologies)]

# 2. Group by 'Country' and sum the 'Value' of the renewable energy technologies
df_sum_renewables = df_renewables.groupby('Country', as_index=False)['F2022'].sum()

# Rename the 'Value' column to 'Total_Renewable'
df_sum_renewables = df_sum_renewables.rename(columns={'F2022': 'Total_Renewable'})


# 3. Filter the DataFrame to get the fossil fuel values
df_fossil_fuels = gc[gc['Technology'] == 'Fossil fuels'][['Country', 'F2022']]



# Rename the 'Value' column to 'Fossil_Fuel'
df_fossil_fuels = df_fossil_fuels.rename(columns={'F2022': 'Fossil_Fuel'})

# 4. Merge the two DataFrames (renewable and fossil fuel) on 'Country'
df_comparison = pd.merge(df_sum_renewables, df_fossil_fuels, on='Country', how='left')

# Display the result
print(df_comparison)
```

```python
# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Melt the DataFrame to plot it easily with seaborn
df_melted = df_comparison.melt(id_vars='Country', value_vars=['Total_Renewable', 'Fossil_Fuel'],
                    var_name='Energy Type', value_name='Value')

# Create a barplot for comparison
sns.barplot(data=df_melted, x='Country', y='Value', hue='Energy Type')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Set labels and title
plt.xlabel('Country')
plt.ylabel('Energy (in GWh)')
plt.title('Comparison of Renewable and Fossil Fuel Energy by Country')

# Display the legend
plt.legend(title='Energy Type')

# Show the plot
plt.tight_layout()
plt.show()
```

### Taking a look at the top 30 countries in terms of energy use and the portion that is renewable

```python
df_comparison['Total_Energy'] = df_comparison['Total_Renewable'] + df_comparison['Fossil_Fuel']

df_sorted = df_comparison.sort_values(by='Total_Energy', ascending=False)

# 3. Select the top 30 
df_top_30 = df_sorted.head(30)

# 5. Plot the stacked bar chart
plt.figure(figsize=(12, 6))

# Plot Fossil Fuel as the base
plt.bar(df_top_30['Country'], df_top_30['Fossil_Fuel'], label='Fossil Fuel', color='brown')

# Plot Renewable energy on top of Fossil Fuel
plt.bar(df_top_30['Country'], df_top_30['Total_Renewable'], bottom=df_top_30['Fossil_Fuel'], label='Renewable', color='green')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Energy (in GWh)')
plt.title('Top 30 Countries by Total Energy Production (Stacked)')

# Show the legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

### Looking at what countries have the highest and lowest amount of renewables

```python
df_percentage = df_sorted
df_percentage['Percent_Renewable'] = df_percentage['Total_Renewable'] / df_percentage['Total_Energy']
df_sorted = df_percentage.sort_values(by=['Percent_Renewable'])



# Sample list of G7 countries
g7_countries = ['United States', 'Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom']
groups = ['Americas', 'Northern Europe', 'Latin America and the Caribbean', 'Southern Europe', 'Western Asia',
          'Oceania sub-regions', 'Advanced Economies', 'G7', 'G20', 'Oceania', 'Northern Africa', 'Cenrtal Asia', 'Africa',
         'Northern America', 'South-eastern Asia', 'Eastern Europe', 'Eastern Asia', 'Sub-Saharan Africa', 'Western Europe',
         'Europe', 'Asia', 'Central Asia']

# Sort by 'Total_Energy' to get the top 40 countries by total energy use
top_50_energy_use = df_sorted.sort_values(by='Total_Energy', ascending=False).head(60)

top_50_energy_use = top_50_energy_use[~top_50_energy_use['Country'].isin(groups)]

# Assign colors based on whether a country is the US, a G7 country, or neither
def assign_color(country):
    if country == 'United States':
        return 'red'  # United States in red
    elif country == 'World':
        return 'purple'
    elif country == 'Emerging and Developing Economies':
        return 'yellow'
    elif country in g7_countries:
        return 'green'  # G7 countries in green
    else:
        return 'blue'  # Other countries in blue

top_50_energy_use['Color'] = top_50_energy_use['Country'].apply(assign_color)

# Sort by percentage renewable energy for better visualization
top_50_energy_use = top_50_energy_use.sort_values(by='Percent_Renewable')

# Create the plot
fig, ax = plt.subplots(figsize=(10, 8))

# Plotting bar chart with colors for the United States, G7, and non-G7 countries
ax.barh(top_50_energy_use['Country'], top_50_energy_use['Percent_Renewable'], color=top_50_energy_use['Color'])

# Add labels and title
ax.set_xlabel('Percentage of Renewable Energy')
ax.set_ylabel('Country')
ax.set_title('Top Countries by Total Energy Use (United States in Red, G7 in Green)')

# Add legend to explain colors
us_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='United States')
emerging_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='purple', markersize=10, label='Emerging and Developing Economies')
world_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='purple', markersize=10, label='World')
g7_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='G7 Country')
non_g7_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Non-G7 Country')
ax.legend(handles=[us_patch, g7_patch, non_g7_patch, world_patch, emerging_patch])

# Show plot
plt.show()
```
